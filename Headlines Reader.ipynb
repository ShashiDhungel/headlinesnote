This is an interactive notebook that reads Nepali news headlines from online sites.


```python
import requests
from bs4 import BeautifulSoup as bs
import re
import nltk
import pandas as pd
```

```python
def soupify (patrika = []):
    
    pagefile = open("C:\\Users\\jp\\AppData\\Local\\Programs\\Python\\Python37-32\\homepages.txt", "r", -1, "utf-8").read()
    
    patrika = [] # Populate the list of URLs to visit for headlines 
    for url in pagefile.splitlines():
        try:
            if "\ufeff" in url:
                patrika.append(str(url.replace("\ufeff", "")))
            else:
                patrika.append(str(url))
        except:
            "Sorry, something seems to have gone wrong. Try again later."
    
    bowl = [] # List of response objects in the bowl
    for paper in patrika:
        try:
            r = requests.get(paper)
            bowl.append(r)
        except:
            "Sorry, something seems to have gone wrong. Try again later."
    
    soup = [] # List of BeautifulSoup objects 
    for r in bowl:
        try:
            spoon = bs(r.text, "html.parser")
            soup.append(spoon)
        except:
            "Sorry, something seems to have gone wrong. Try again later."
    return soup
```

```python
mixed_soup = soupify(patrika = [])
```

```python
print("We now have a {} of {} news sites of Nepal.".format(type(mixed_soup), str(len(mixed_soup))))
```

```python
outlets = []
counter = 0
for soup in mixed_soup:
    print(counter + 1, soup.find("title"))
    outlets.append(soup.find("title"))
    counter += 1
```

```python
heads = []
for soup in mixed_soup:
    headlines = soup.find_all(re.compile("h[1-6]"))
    for line in headlines:
        if line.string not in heads:
            heads.append(line.string)
            for link in line.find_all(re.compile("^a")):
                if "http" not in str(link):
                    if "news/2017-" in str(link):
                        heads.append("http://kantipur.ekantipur.com" + str(link.get("href")))
                    elif "/news/2" in str(link):
                        heads.append("http://www.nagariknews.com" + str(link.get("href")))                     
                    elif "/story/" in str(link):
                        heads.append("http://www.ratopati.com" + str(link.get("href")))
                else:
                    heads.append(str(link.get("href")))
```

```python
headlinks = []
for h in heads:
    if h not in headlinks:
        headlinks.append(str(h))
```

```python
for h in headlinks:
    print(str(h).lstrip())
```

```python
words = []
for head in headlines:
    for word in str(head).split(" "):
        words.append(word)
```

```python
fdist = nltk.FreqDist(words)
```

```python
x = fdist.hapaxes()
```

```python
newhap = []
i = 0
for eachword in x:
    st = str(x[i]).replace("[,]", "")
    if st not in newhap:
        newhap.append(st)
        i = i + 1
```
